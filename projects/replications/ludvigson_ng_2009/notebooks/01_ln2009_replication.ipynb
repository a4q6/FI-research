# Ludvigson & Ng (2009) 再現

> Ludvigson, S. C., & Ng, S. (2009). Macro Factors in Bond Risk Premia. *RFS*, 22(12), 5027-5067.

詳細は [README.md](../README.md) を参照。本ノートブックは結果の表示用。

**核心**: マクロ PCA 因子は CP factor を超える追加情報を持つ (+9〜+14 ポイント ΔR²)。ポスト QE 期 (2008-2025) では macro factors の方が CP factor より強い予測力。

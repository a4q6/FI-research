# Gilchrist & Zakrajšek (2012) 再現

> Gilchrist, S., & Zakrajšek, E. (2012). Credit Spreads and Business Cycle Fluctuations. *AER*, 102(4), 1692-1720.

詳細は [README.md](../README.md) を参照。本ノートブックは結果の表示用。

**核心**: 論文期間 (1973-2010) で EBP は IP / Payroll / UR を強く予測 (R² 25-60%, t-stat 5-10)。**論文後 (2011-2025) で予測力が完全に崩壊** (R² ≈ 0, 全係数 insignificant) → QE による構造変化を示唆。

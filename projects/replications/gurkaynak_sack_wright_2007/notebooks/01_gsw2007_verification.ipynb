# Gürkaynak, Sack & Wright (2007) 検証

> Gürkaynak, R. S., Sack, B., & Wright, J. H. (2007). The U.S. Treasury yield curve: 1961 to the present. *JME*, 54(8), 2291-2304.

詳細は [README.md](../README.md) を参照。本ノートブックは結果の表示用。

**核心**: Svensson 公式と公開 SVENY は **数値誤差レベル (RMS 0.002 bps, max 0.04 bps)** で完全一致 → GSW データの内部整合性は完璧。FRED CMT との差は 5-15 bps の 系統的パターン（複利規約 + on-the-run プレミアム）。
